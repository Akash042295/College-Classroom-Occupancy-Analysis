#College Classroom Occupancy Analysis

▪From: Minor 1 – Pandas.

▪Extension: Analyze classroom usage → space optimization.

▪Add-ons: Bar charts, summary stats.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Room_No.': ['A101', 'A102', 'B201', 'B202', 'C301', 'C302'],
    'Capacity': [100, 30, 50, 200, 40, 120],
    'Total_Student': [85, 28, 15, 190, 45, 30],
    'Course': ['Math', 'Physics', 'Chemistry', 'English', 'History', 'Art'],
    'start_Time': ['9:00 AM', '9:00 AM', '11:00 AM', '11:00 AM', '1:00 PM', '1:00 PM'],
    'end_Time': ['11:00 AM', '11:00 AM', '1:00 PM', '1:00 PM', '3:00 PM', '3:00 PM']
}
df = pd.DataFrame(data)
df
```

```python
df['Utilization'] = (df['Total_Student']/ df['Capacity'])*100
df
```

```python
df.groupby('Room_No.')['Utilization'].mean().plot(kind='bar')
plt.xlabel('Room_No.')
plt.ylabel('Avg Utilization %')
plt.title('Classroom utilization')
plt.show()
```
